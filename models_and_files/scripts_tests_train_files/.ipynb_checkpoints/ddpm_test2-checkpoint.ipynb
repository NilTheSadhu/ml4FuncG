```python
from torch import nn
import torch.nn.functional as F
import torch    
import numpy as np
import pandas as pd
from torch.utils.data import DataLoader
from torch.optim import Adam
from tasks.spatial_transcriptomics_dataset import MemoryEfficientPairedLoader, MemoryEfficientPairedLoaderWrapper, PairedSpatialTranscriptomicsDataset
```

```python
def align_tensor_sizes(tensor_a, tensor_b):
    """Ensure tensor_a matches the dimensions of tensor_b."""
    _, _, height_t1, width_t1 = tensor_a.size()
    _, _, height_t2, width_t2 = tensor_b.size()

    if height_t1 != height_t2 or width_t1 != width_t2:
        tensor_a = F.interpolate(tensor_a, size=(height_t2, width_t2), mode='bilinear', align_corners=False)

    return tensor_a

class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels,kernel_size=5, use_attention=False):
        super(ResidualBlock, self).__init__()
        self.use_attention = use_attention

        # Main convolutional path
        self.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1)
        self.norm1 = nn.GroupNorm(8, out_channels)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1)
        self.norm2 = nn.GroupNorm(8, out_channels)

        # Residual projection for matching dimensions
        self.residual_projection = (
            nn.Conv2d(in_channels, out_channels, kernel_size=1)
            if in_channels != out_channels
            else nn.Identity()
        )

    def forward(self, x):
        residual = self.residual_projection(x)
        x = F.relu(self.norm1(self.conv1(x)))
        x = self.norm2(self.conv2(x))
        return F.relu(x + residual)

class UNet(nn.Module):
    def __init__(self, in_channels=502, out_channels=502, base_channels=128):
        super(UNet, self).__init__()

        # Use kernel_size=5 in the first encoder layers to match the 5x5 structure
        self.encoder1 = ResidualBlock(in_channels, base_channels, kernel_size=5)
        self.encoder2 = ResidualBlock(base_channels, base_channels * 2, kernel_size=3)
        self.encoder3 = ResidualBlock(base_channels * 2, base_channels * 4, kernel_size=3)
        self.encoder4 = ResidualBlock(base_channels * 4, base_channels * 8, kernel_size=3)

        # Bottleneck
        self.middle = ResidualBlock(base_channels * 8, base_channels * 8, kernel_size=3, use_attention=True)

        # Decoder
        self.decoder4 = nn.ConvTranspose2d(base_channels * 8, base_channels * 4, kernel_size=4, stride=2, padding=1)
        self.decoder3 = nn.ConvTranspose2d(base_channels * 4, base_channels * 2, kernel_size=4, stride=2, padding=1)
        self.decoder2 = nn.ConvTranspose2d(base_channels * 2, base_channels, kernel_size=4, stride=2, padding=1)
        self.decoder1 = nn.ConvTranspose2d(base_channels, out_channels, kernel_size=4, stride=2, padding=1)

        # Skip connections
        self.skip4 = nn.Conv2d(base_channels * 4, base_channels * 4, kernel_size=1)
        self.skip3 = nn.Conv2d(base_channels * 2, base_channels * 2, kernel_size=1)
        self.skip2 = nn.Conv2d(base_channels, base_channels, kernel_size=1)


    def forward(self, x, original_size):
        #encoder with skip cons
        enc1 = self.encoder1(x)
        enc2 = self.encoder2(F.avg_pool2d(enc1, kernel_size=2))
        enc3 = self.encoder3(F.avg_pool2d(enc2, kernel_size=2))
        enc4 = self.encoder4(F.avg_pool2d(enc3, kernel_size=2))

        # Bottleneck
        mid = self.middle(enc4)

        #Decoder with aligned skip connections - alignment/dim adjust
        dec4 = self.decoder4(mid)
        dec4 = align_tensor_sizes(dec4, self.skip4(enc3)) + self.skip4(enc3)

        dec3 = self.decoder3(dec4)
        dec3 = align_tensor_sizes(dec3, self.skip3(enc2)) + self.skip3(enc2)

        dec2 = self.decoder2(dec3)
        dec2 = align_tensor_sizes(dec2, self.skip2(enc1)) + self.skip2(enc1)

        dec1 = self.decoder1(dec2)

        #add crop in the end just in case
        return dec1[:, :, :original_size[0], :original_size[1]]



def masked_loss(output, target, padding_value=-1):
    #Calculate the loss while ignoring padded regions.
    mask = target != padding_value  # Mask out padding values
    loss = F.mse_loss(output[mask], target[mask], reduction="mean")
    return loss
# === Diffusion Model with Adjustments === #
class DDPM(nn.Module):
    def __init__(self, image_size, channels, timesteps=1000):
        super(DDPM, self).__init__()
        self.timesteps = timesteps
        self.image_size = image_size

        #attempt to use unet directly
        self.unet = UNet(in_channels=channels, out_channels=channels)

    def forward(self, x, t, original_size):
        #pass for unet
        return self.unet(x, original_size)


class DiffusionTraining:
    def __init__(self, model, timesteps=1000, beta_start=0.0001, beta_end=0.02):
        self.model = model
        self.timesteps = timesteps

        #beta schedule definition - begin with linear 
        self.betas = torch.linspace(beta_start, beta_end, timesteps)
        self.alphas = 1.0 - self.betas
        self.alpha_bars = torch.cumprod(self.alphas, dim=0)

    def forward_diffusion(self, x_0, t):
        """
        Apply the forward diffusion process to input x_0 at timestep t.
        """
        noise = torch.randn_like(x_0)
        alpha_bar_t = self.alpha_bars.to(t.device)[t].view(-1, 1, 1, 1)
        return torch.sqrt(alpha_bar_t) * x_0 + torch.sqrt(1 - alpha_bar_t) * noise, noise

    def train_step(self, x_0, optimizer, original_size):
        """
        Perform a single training step.
        """
        t = torch.randint(0, self.timesteps, (x_0.size(0),), device=x_0.device)
        x_t, noise = self.forward_diffusion(x_0, t)

        # Forward pass through the model
        noise_pred = self.model(x_t, t, original_size)

        # Compute loss using masking
        loss = masked_loss(noise_pred, noise)

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        return loss.item()

#initialize model for testing
image_size = 313
channels = 502
model = DDPM(image_size, channels)
print('done')
device = 'cuda' #no if just cuda - do or do not, there is not try
image_size = 313
input_channels = 502
model = DDPM(image_size, input_channels).to(device)
trainer = DiffusionTraining(model)
optimizer = Adam(model.parameters(), lr=3e-4)
```

```python


#file paths, feel free to change range and number if you want to cut it down for quick training/testing
train_high_res_files = [f"binned_data/high_res_grids_{i}.npz" for i in range(4, 25)]
train_low_res_files = [f"binned_data/low_res_grids_{i}.npz" for i in range(4, 25)]
val_high_res_files = [f"binned_data/high_res_grids_{i}.npz" for i in range(1, 4)]
val_low_res_files = [f"binned_data/low_res_grids_{i}.npz" for i in range(1, 4)]
test_high_res_files = [f"binned_data/low_res_grids_{i}.npz" for i in range(25, 28)]
test_low_res_files = [f"binned_data/low_res_grids_{i}.npz" for i in range(25, 28)]

#___# <- thats my face - This implementation pre-dates my use of the wrappers - you can update this to use the wrapper if you want
print("creating loaders")
train_loader = MemoryEfficientPairedLoader(train_high_res_files, train_low_res_files, batch_size=10)
val_loader = MemoryEfficientPairedLoader(val_high_res_files, val_low_res_files, batch_size=10)
test_loader = MemoryEfficientPairedLoader(test_high_res_files, test_low_res_files, batch_size=10)

checkpoint_path = 'checkpoints/ddpm4_checkpoint.pth' #change this file name if you don't want to overwrite files and plan to run this same block
#you'll be surprised how many files I lost from overwriting :/
def save_checkpoint(epoch, model, optimizer, train_losses, val_losses, checkpoint_path):
    torch.save({
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'train_losses': train_losses,
        'val_losses': val_losses,
    }, checkpoint_path)

#beginning training - 50 to 100 epochs is what is suggested. 20 is what i've seen medical papers use with diffusion SR of images. This is early testing so 15 with possible early termination depending on training progress
num_epochs = 15
train_losses, val_losses = [], []

for epoch in range(num_epochs):
    print(f"Epoch {epoch + 1}/{num_epochs}")
    model.train()
    epoch_train_loss = 0

    for batch in train_loader:
        lr_data = batch['low_res_tensor'].to(device)
        hr_data = batch['high_res_tensor'].to(device)
        mask = batch['mask'].to(device)

        # Forward diffusion
        t = torch.randint(0, trainer.timesteps, (lr_data.size(0),), device=device)
        x_t, noise = trainer.forward_diffusion(hr_data, t)

        # Predict noise
        noise_pred = model(x_t, t, original_size=(313, 313))

        # Align dimensions explicitly
        noise_pred = align_tensor_sizes(noise_pred, mask)

        # Masked loss
        loss = F.mse_loss(noise_pred * mask, noise * mask, reduction='sum') / mask.sum()

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        epoch_train_loss += loss.item()

    avg_train_loss = epoch_train_loss / len(train_loader)
    train_losses.append(avg_train_loss)
    print(f"Train Loss: {avg_train_loss:.4f}")

    # Validation loop
    model.eval()
    epoch_val_loss = 0
    with torch.no_grad():
        for batch in val_loader:
            lr_data = batch['low_res_tensor'].to(device)
            hr_data = batch['high_res_tensor'].to(device)
            mask = batch['mask'].to(device)

            t = torch.randint(0, trainer.timesteps, (lr_data.size(0),), device=device)
            x_t, noise = trainer.forward_diffusion(hr_data, t)
            noise_pred = model(x_t, t, original_size=(313, 313))

            #explicit dim align 
            noise_pred = align_tensor_sizes(noise_pred, mask)

            #masked validation loss
            val_loss = F.mse_loss(noise_pred * mask, noise * mask, reduction='sum') / mask.sum()
            epoch_val_loss += val_loss.item()

    avg_val_loss = epoch_val_loss / len(val_loader)
    val_losses.append(avg_val_loss)
    print(f"Validation Loss: {avg_val_loss:.4f}")
    save_checkpoint(epoch, model, optimizer, train_losses, val_losses, checkpoint_path)

#Test evaluation
model.eval()
epoch_test_loss = 0
with torch.no_grad():
    for batch in test_loader:
        lr_data = batch['low_res_tensor'].to(device)
        hr_data = batch['high_res_tensor'].to(device)
        mask = batch['mask'].to(device)

        t = torch.randint(0, trainer.timesteps, (lr_data.size(0),), device=device)
        x_t, noise = trainer.forward_diffusion(hr_data, t)
        noise_pred = model(x_t, t, original_size=(313, 313))

        print("explicit dimension correction")
        noise_pred = align_tensor_sizes(noise_pred, mask)

        print("masked test loss calc")
        test_loss = F.mse_loss(noise_pred * mask, noise * mask, reduction='sum') / mask.sum()
        epoch_test_loss += test_loss.item()

avg_test_loss = epoch_test_loss / len(test_loader)
print(f"Test Loss: {avg_test_loss:.4f}")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from skimage.metrics import structural_similarity as ssim
import numpy as np

def align_tensor_sizes(tensor_one, tensor_two):
    #typically we won't have to use this but this is a precaution for slight mismatches in dims as we propogate especially given our odd numbered starting dim m
    trash, trash, height_t1, width_t1 = tensor_one.size()
    trash, trash, height_t2, width_t2 = tensor_one.size()

    if height_t1 != height_t2 or width_t1 != width_t2:
        tensor_one = F.interpolate(tensor_one, size=(height_t2, width_t2), mode='bilinear', align_corners=False)

    return tensor_a


class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=5, use_attention=True):
        super(ResidualBlock, self).__init__()
        self.use_attention = use_attention

        self.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=kernel_size, padding=1)
        self.norm1 = nn.GroupNorm(8, out_channels)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=kernel_size, padding=1)
        self.norm2 = nn.GroupNorm(8, out_channels)

        #this projection should typically not be substantial if ever used - tuned layers to preserve final dims
        self.residual_projection = (
            nn.Conv2d(in_channels, out_channels, kernel_size=1)
            if in_channels != out_channels
            else nn.Identity()
        )

    def forward(self, x):
        residual = self.residual_projection(x)
        x = F.mish(self.norm1(self.conv1(x)))
        x = self.norm2(self.conv2(x))
        #precautionary step - debug to make sure you aren't dramatically resizing ever otherwise thats a problem
        residual = align_tensor_sizes(residual, x)
        
        return F.mish(x + residual)

class UNet(nn.Module):
    def __init__(self, in_channels=512, out_channels=512, base_channels=256):
        super(UNet, self).__init__()

        #ported from before
        self.encoder1 = ResidualBlock(in_channels, base_channels, kernel_size=5)
        self.encoder2 = ResidualBlock(base_channels, base_channels * 2, kernel_size=3)
        self.encoder3 = ResidualBlock(base_channels * 2, base_channels * 4, kernel_size=3)
        self.encoder4 = ResidualBlock(base_channels * 4, base_channels * 8, kernel_size=3)

        #bottleneck for global spatial detail capture with attention
        self.middle = ResidualBlock(base_channels * 8, base_channels * 8, kernel_size=3, use_attention=True)

        #decoder - symmetry check
        self.decoder4 = nn.ConvTranspose2d(base_channels * 8, base_channels * 4, kernel_size=4, stride=2, padding=1)
        self.decoder3 = nn.ConvTranspose2d(base_channels * 4, base_channels * 2, kernel_size=4, stride=2, padding=1)
        self.decoder2 = nn.ConvTranspose2d(base_channels * 2, base_channels, kernel_size=4, stride=2, padding=1)
        self.decoder1 = nn.ConvTranspose2d(base_channels, out_channels, kernel_size=4, stride=2, padding=1)

        #implement skip connections
        self.skip4 = nn.Conv2d(base_channels * 4, base_channels * 4, kernel_size=1)
        self.skip3 = nn.Conv2d(base_channels * 2, base_channels * 2, kernel_size=1)
        self.skip2 = nn.Conv2d(base_channels, base_channels, kernel_size=1)

    def forward(self, x, original_size):
        # Encoder
        enc1 = self.encoder1(x)
        enc2 = self.encoder2(F.avg_pool2d(enc1, kernel_size=2))
        enc3 = self.encoder3(F.avg_pool2d(enc2, kernel_size=2))
        enc4 = self.encoder4(F.avg_pool2d(enc3, kernel_size=2))

        #to bottle
        mid = self.middle(enc4)

        #decoder with aligned skip connections
        dec4 = self.decoder4(mid)
        dec4 = align_tensor_sizes(dec4, self.skip4(enc3)) + self.skip4(enc3)

        dec3 = self.decoder3(dec4)
        dec3 = align_tensor_sizes(dec3, self.skip3(enc2)) + self.skip3(enc2)

        dec2 = self.decoder2(dec3)
        dec2 = align_tensor_sizes(dec2, self.skip2(enc1)) + self.skip2(enc1)

        dec1 = self.decoder1(dec2)

        return dec1[:, :, :original_size[0], :original_size[1]]

#preliminary weighted loss function -> refined in SRCode
def weighted_loss(output, target, padding_value=-1):
    """
    Calculate a weighted loss combining MSE and SSIM.
    """
    mask = target != padding_value  # Mask out padding values
    mse_loss = F.mse_loss(output[mask], target[mask], reduction="mean")

    # Compute SSIM for each channel and average over channels
    output_np = output[mask].detach().cpu().numpy()
    target_np = target[mask].detach().cpu().numpy()

    ssim_values = []
    for c in range(output_np.shape[1]):  # Loop over channels
        ssim_channel = ssim(output_np[c], target_np[c], data_range=target_np.max() - target_np.min())
        ssim_values.append(ssim_channel)
    
    avg_ssim = np.mean(ssim_values)
    ssim_loss = 1 - avg_ssim

    # Weighted combination
    return 0.8 * mse_loss + 0.2 * ssim_loss

class DDPM(nn.Module):
    def __init__(self, image_size, channels, timesteps=1000):
        super(DDPM, self).__init__()
        self.timesteps = timesteps
        self.image_size = image_size

        # Use UNet directly
        self.unet = UNet(in_channels=channels, out_channels=channels)

    def forward(self, x, t, original_size):
        # Forward pass through U-Net
        output = self.unet(x, original_size)
        
        # Align output dimensions to match input size if necessary
        output = align_tensor_sizes(output, x)
        
        return output

#attempting with cosine schedule
class DiffusionTraining:
    def __init__(self, model, timesteps=1000, beta_start=0.0001, beta_end=0.02,schedule="cosine"):
        self.model = model
        self.timesteps = timesteps
        self.schedule=schedule
        if self.schedule == 'linear':
            self.betas = torch.linspace(beta_start, beta_end, timesteps)
        elif self.schedule == 'cosine':
            self.betas = self._cosine_beta_schedule(timesteps)
        else:
            raise ValueError("Unsupported schedule type. Use 'linear' or 'cosine'.")

        self.alphas = 1.0 - self.betas
        self.alpha_bars = torch.cumprod(self.alphas, dim=0)

    def _cosine_beta_schedule(self, timesteps, s=1e-5):
    
        #Generate beta values based on a cosine schedule.

        steps = torch.linspace(0, timesteps, timesteps + 1) / timesteps  # Normalize timesteps
        alphas_cumprod = torch.cos((steps * (1 + s) + s) * torch.pi / 2) ** 2
        alphas_cumprod = alphas_cumprod / alphas_cumprod[0]  # Normalize to start at 1

        # Convert alpha bars (\bar{\alpha}) to betas
        betas = 1 - (alphas_cumprod[1:] / alphas_cumprod[:-1])
        betas = torch.clamp(betas, min=1e-8, max=0.999)  # Avoid instability
        return betas
    def forward_diffusion(self, x_0, t):
        """
        Apply the forward diffusion process to input x_0 at timestep t.
        """
        noise = torch.randn_like(x_0)
        alpha_bar_t = self.alpha_bars.to(t.device)[t].view(-1, 1, 1, 1)
        return torch.sqrt(alpha_bar_t) * x_0 + torch.sqrt(1 - alpha_bar_t) * noise, noise

    def train_step(self, x_0, optimizer, original_size):
        """
        Perform a single training step.
        """
        t = torch.randint(0, self.timesteps, (x_0.size(0),), device=x_0.device)
        x_t, noise = self.forward_diffusion(x_0, t)

        # Forward pass through the model and compute weighted loss
        noise_pred = self.model(x_t, t, original_size)
        loss = weighted_loss(noise_pred, noise)

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        return loss.item()
print("done")
def align_tensor_sizes(tensor_a, tensor_b):
    """Ensure tensor_a matches the dimensions of tensor_b."""
    _, _, height_t1, width_t1 = tensor_a.size()
    _, _, height_t2, width_t2 = tensor_b.size()

    if height_t1 != height_t2 or width_t1 != width_t2:
        tensor_a = F.interpolate(tensor_a, size=(height_t2, width_t2), mode='bilinear', align_corners=False)
    return tensor_a

class DDPM(nn.Module):
    def __init__(self, image_size, channels, timesteps=1000):
        super(DDPM, self).__init__()
        self.timesteps = timesteps
        self.image_size = image_size

        # Use UNet directly
        self.unet = UNet(in_channels=channels, out_channels=channels)

    def forward(self, x, t, original_size):
        # Forward pass through U-Net
        return self.unet(x, original_size)



device = 'cuda' if torch.cuda.is_available() else 'cpu'
image_size = 313
input_channels = 502
model = DDPM(image_size, input_channels).to(device)
trainer = DiffusionTraining(model)
optimizer = Adam(model.parameters(), lr=1e-3)
```

```python
#change error function
def hybrid_ssim_mse(predictions, targets, alpha=0.4, beta=0.8, weight_factor=30):
    
    #WE want to weight the spatial coordinates more in loss calculations as well as incorporate SSIM
    #SSIM should have a lower weight as we have effectively created "soft bounds" by utilizing a spatially encoded 
    #grid with coordinates embedded as features
    #predictions and targets must be on CPU and detached for NumPy conversion!!
    #ssim module we are using requires it all to be numpy
    predictions_np = predictions.squeeze(0).detach().cpu().numpy()
    targets_np = targets.squeeze(0).detach().cpu().numpy()

    #IMPORTANT NOTE: multichannel SSMI expects (H, W, C) format!!! 
    print(predictions_np.shape,targets_np.shape) #debugging
    predictions_np = predictions_np.transpose(1, 2, 0)  #hwc
    targets_np = targets_np.transpose(1, 2, 0)          #hwc format

    #multichannel since we have multiple channels/features (genes/spatial coords) - I wish I knew of this before
    ssim_score = ssim(
        predictions_np,
        targets_np,
        data_range=targets_np.max() - targets_np.min(),
        multichannel=True
    )
    print("ssim: ",ssim_score) 
    #we want to add some weight to the penalty for misrepresenting spatial coordinates due to their scale
    mse = (predictions - targets) ** 2
    weights = torch.ones_like(predictions)
    weights[:, -2:, :, :] *= weight_factor  #last two channels are x and y
    weighted_mse = torch.mean(mse * weights)
    hybrid_loss = alpha * (1 - ssim_score) + beta * weighted_mse #80% MSE and 20% SSM since soft bounds are effectively in existence within our grid
    return hybrid_loss
```
